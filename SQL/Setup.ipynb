```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
from sqlalchemy import text
```

```python
data = pd.read_csv("Spring 2023.csv", sep=",")
data
```

```python
data = data[pd.to_numeric(data["Catalog Nbr"], errors='coerce').notnull()]
data
```

```python
data.columns[3:21]
```

```python
data.drop(columns=data.columns[3:21], inplace=True)
```

```python
data.fillna("EMPTY")
```

```python
def collateInstructors(row):
    ret = row.iloc[0]
    i = 1
    while((i < 6) and (row.iloc[i] != "EMPTY")):
        ret = ret + '; ' + row.iloc[i]
        i = i + 1
    return ret
```

```python
data["Instructors"] = (data.apply( lambda x: collateInstructors(x[3:9]), axis=1))
```

```python
data.drop(columns=data.columns[3:9], inplace=True)
data
```
